```python
def get_url_params(url):
    params = {}
    parts = url.split('?')
    if len(parts) > 1:
        query_string = parts[1]
        for param in query_string.split('&'):
            key_value = param.split('=')
            if len(key_value) == 2:
                params[key_value[0]] = key_value[1]
    return params

def extract_data(html):
    soup = BeautifulSoup(html, 'html.parser')
    # 根据网页实际结构提取所需数据，这里只是一个示例占位
    data = []
    # 假设文档列表在一个特定的类名的元素中
    document_list = soup.find_all('div', class_='news_list')
    for item in document_list:
        title = item.find('h3', class_='title').text if item.find('h3', class_='title') else None
        description = item.find('p', class_='description').text if item.find('p', class_='description') else None
        data.append({'title': title, 'description': description})
    return data

url = "http://www.bjrd.gov.cn/so/wjsjk/s?tab=all&siteCode=1100000088fgk&qt="
response = requests.get(url)

params = get_url_params(url)
data = extract_data(response.content)
print(data)
for item in data:
    print(f"Title: {item['title']}\nDescription: {item['description']}\n")
```
