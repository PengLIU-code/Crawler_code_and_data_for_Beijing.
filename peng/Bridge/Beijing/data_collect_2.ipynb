```python
import requests
from bs4 import BeautifulSoup

# 获取网页内容
url = 'https://www.beijing.gov.cn/zhengce/gfxwj/sj/202405/t20240527_3694615.html'
response = requests.get(url)
html_content = response.content

soup = BeautifulSoup(html_content, 'html.parser')

# 政策名称
policy_title = soup.h1.p.text.strip() if soup.h1 and soup.h1.p else 'N/A'
print(f"政策名称：{policy_title}")

# 发文机构
issuing_agency_tag = soup.find('meta', attrs={'name': 'ContentSource'})
issuing_agency = issuing_agency_tag['content'] if issuing_agency_tag else 'N/A'
print(f"发文机构：{issuing_agency}")

# 成文日期
date_formed_tag = soup.find('li', string='[成文日期]')
date_formed = date_formed_tag.find_next_sibling('li').span.text.strip() if date_formed_tag else 'N/A'
print(f"成文日期：{date_formed}")

# 发布日期
date_published_tag = soup.find('meta', attrs={'name': 'PubDate'})
date_published = date_published_tag['content'] if date_published_tag else 'N/A'
print(f"发布日期：{date_published}")

# 主题分类
topic_category_tag = soup.find('li', string='[主题分类]')
topic_category = topic_category_tag.find_next_sibling('li').span.text.strip() if topic_category_tag else 'N/A'
print(f"主题分类：{topic_category}")

# 有效性
validity_tag = soup.find('li', string='[有效性]')
validity = validity_tag.find_next_sibling('li').span.text.strip() if validity_tag else 'N/A'
print(f"有效性：{validity}")

# 文件来源
file_source_tag = soup.find('li', string='[文件来源]')
file_source = file_source_tag.find_next_sibling('li').span.text.strip() if file_source_tag else 'N/A'
print(f"文件来源：{file_source}")

# 发文字号
document_number = ""
fwzhdz = soup.find('span', class_='fwzhdz')
fwzhnf = soup.find('span', class_='fwzhnf')
fwzhxh = soup.find('span', class_='fwzhxh')
if fwzhdz and fwzhnf and fwzhxh:
    document_number = f"{fwzhdz.text.strip()}{fwzhnf.text.strip()}{fwzhxh.text.strip()}"
print(f"发文字号：{document_number}")
```

```python

title_info = soup.find('div', class_='container').find('ol')
print(title_info)

# 提取结构信息
def extract_info(label):
    tag = title_info.find('li', string=lambda x: x and label in x)
    return tag.find('span').text.strip() if tag else 'N/A'

theme = extract_info('[主题分类]')
issuing_agency = extract_info('[发文机构]')
joint_issuing_agency = extract_info('[联合发文单位]')
implementation_date = extract_info('[实施日期]')
date_formed = extract_info('[成文日期]')
abolition_date = extract_info('[废止日期]')
date_published = extract_info('[发布日期]')
validity = extract_info('[有效性]')
file_source_tag = title_info.find('li', string=lambda x: x and '[文件来源]' in x)
file_source = file_source_tag.find('span').text.strip() if file_source_tag else 'N/A'

# 发文字号
document_number = ""
fwzhdz = title_info.find('span', class_='fwzhdz')
fwzhnf = title_info.find('span', class_='fwzhnf')
fwzhxh = title_info.find('span', class_='fwzhxh')
if fwzhdz and fwzhnf and fwzhxh:
    document_number = f"{fwzhdz.text.strip()}{fwzhnf.text.strip()}{fwzhxh.text.strip()}"

# 打印提取的信息
print(f"主题分类：{theme}")
print(f"发文机构：{issuing_agency}")
print(f"联合发文单位：{joint_issuing_agency}")
print(f"实施日期：{implementation_date}")
print(f"成文日期：{date_formed}")
print(f"废止日期：{abolition_date}")
print(f"发布日期：{date_published}")
print(f"有效性：{validity}")
print(f"文件来源：{file_source}")
print(f"发文字号：{document_number}")
```
